```python
import rofa
from rofa import QuantileSimulator
```

```python
import pandas as pd
```

```python
# Register Returns Data
returns = pd.read_pickle('data/returns.pkl')
rofa.register_returns(returns)
```

```python
returns.head()
```

```python
# Read Close Data
close = pd.read_pickle('data/close.pkl')
```

```python
close.head()
```

```python
# Create QuantileSimulator instance
quan_sim = QuantileSimulator(close, rebalance_freq=10, bins=5)
```

```python
# Run simulation
quan_sim.run()
```

```python
# Plot Portfolio Returns
quan_sim.plotter.plot_portfolio_returns()
```

```python
# Plot Performance Metrics
quan_sim.plotter.plot_performance_metrics()
```

```python
# Plot Rolling Performance
quan_sim.plotter.plot_rolling_performance()
```

```python
# Plot All
quan_sim.plotter.plot_all()
```
